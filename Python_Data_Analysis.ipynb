# 5.0 Data Cleaning

```
# Map the lowering function to all column names
df1 = pd.read_csv('Telcom_dataset.csv')
df2 = pd.read_csv('Telcom_dataset2.csv')
df3 = pd.read_csv('Telcom_dataset3.csv')

df1.columns = map(str.lower, df1.columns)
df1
df2.columns = map(str.lower, df2.columns)
df2
df3.columns = map(str.lower, df3.columns)
df3

df1[:2]
df2[:2]
df3[:2]
```

```
# To remove duplicates from the datasets
# Identify which observations are duplicates

df1.duplicated()

# Drop duplicates

non_duplicated_df1 = df1.drop_duplicates()
df_1 = pd.DataFrame(non_duplicated_df1)
df_1
```

```
# To remove duplicates from the datasets
# Identify which observations are duplicates

df2.duplicated()

# Drop duplicates

non_duplicated_df2 = df2.drop_duplicates()
df_2 = pd.DataFrame(non_duplicated_df2)
df_2
```

```
# To remove duplicates from the datasets
# Identify which observations are duplicates

df3.duplicated()

# Drop duplicates

non_duplicated_df3 = df3.drop_duplicates()
df_3 = pd.DataFrame(non_duplicated_df3)
df_3
```

```
# Checking if there are missing values in my dataset

df_1.isnull() 


# Drop missing observations

df_1_no_missing = df_1.dropna()
df_1_no_missing
```

```
# To drop rows where all cells in that row is NA

df_2_cleaned = df_2.dropna(how='all')
df_2_cleaned

# To drop column if they only contain missing values

df_2.dropna(axis=1, how='all')

# To drop rows that contain less than five observations

df_2.dropna(thresh=2)

# To fill in missing data with zeros

df_2.fillna(0)
```
